```python
def parsefile(fname):
    filepath = 'n32R8.rpt'
    with open(filepath) as fp:
        line = fp.readline()
        cnt = 29
        for x in range(0, 30):
            line = fp.readline()
        while line:

            if re.search("1. CLB Logic",line):
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,12):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()
            elif re.search("2. BLOCKRAM", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,4):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()
            elif re.search("3. ARITHMETIC", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,1):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()
            elif re.search("4. I/O", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,1):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()     

            elif re.search("5. CLOCK", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,7):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()

            elif re.search("6. ADVANCED", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,12):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()

            elif re.search("7. CONFIGURATION", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,7):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()

            elif re.search("8. Primitives", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,7):
                    temp = line
                    temp = temp.split("|")
                    Refname = temp[1].strip()
                    used = temp[2].strip()
                    functional_cat = temp[3].strip()
                    int(used)
                    print("Ref Name: " + Refname + ", used:" + used + ", Functional Category: " + functional_cat)
                    line = fp.readline()

            elif re.search("11. SLR Connectivity", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,18):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    used = temp[2].strip()
                    fixed = temp[3].strip()
                    available = temp[4].strip()
                    util = temp[5].strip()
                    if used != '':
                        int(used)
                    if fixed != '':
                        int(fixed)
                    if available != '':
                        int(available)
                    if util != '':
                        float(util)
                    print("Site Type: " + site_type + ", used:" + used + ", fixed: " + fixed + ", Available: " + available + ", Util: " + util)
                    line = fp.readline()
            elif re.search("12. SLR Connectivity Matrix", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,3):
                    temp = line
                    temp = temp.split("|")
                    from_to = temp[1].strip()
                    slr2 = temp[2].strip()
                    slr1 = temp[3].strip()
                    slr0 = temp[4].strip()
                    if slr2 != '':
                        int(slr2)
                    if slr1 != '':
                        int(slr1)
                    if slr0 != '':
                        int(slr0)
                    print("FROM \ TO " + from_to + ", SLR2:" + slr2 + ", SLR1: " + slr1 + ", SLR0: " + slr0)
                    line = fp.readline() 

            elif re.search("13. SLR CLB Logic and Dedicated Block Utilization", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,21):
                    temp = line
                    temp = temp.split("|")
                    site_type = temp[1].strip()
                    slr0 = temp[2].strip()
                    slr1 = temp[3].strip()
                    slr2 = temp[4].strip()
                    slr0percent = temp[5].strip()
                    slr1percent = temp[6].strip()
                    slr2percent = temp[7].strip()
                    if slr0 != '':
                        int(slr0)
                    if slr1 != '':
                        int(slr1)
                    if slr2 != '':
                        int(slr2)
                    if slr0percent != '':
                        float(slr0percent)
                    if slr1percent != '':
                        float(slr1percent)
                    if slr2percent != '':
                        float(slr2percent)
                    print("Site Type: " + site_type + ", SLR0:" + slr0 + ", SLR1: " + slr1 + ", SLR2: " + slr2 + ", SLR0%: " + slr0percent + ", SLR1%: " + slr1percent + ", SLR2%: " + slr2percent)
                    line = fp.readline()
            elif re.search("14. SLR IO Utilization", line):
                print()
                print(line)
                for x in range(0 , 6):
                    line = fp.readline()
                for g in range(0,4):
                    temp = line
                    temp = temp.split("|")
                    SLR_index = temp[1].strip()
                    used_IOB = temp[2].strip()
                    percent_IOB = temp[3].strip()
                    used_ipads = temp[4].strip()
                    percent_ipads = temp[5].strip()
                    used_opads = temp[6].strip()
                    percent_opads = temp[7].strip()
                    gts = temp[8].strip()
                    if used_IOB != '':
                        int(used_IOB)
                    if percent_IOB != '':
                        float(percent_IOB)
                    if used_ipads != '':
                        int(used_ipads)
                    if percent_ipads != '':
                        float(percent_ipads)
                    if used_opads != '':
                        int(used_opads)
                    if percent_opads != '':
                        float(percent_opads)
                    if gts != '':
                        int(gts)
                    print("SLR Index: " + SLR_index + ", USED IOBs:" + used_IOB + ", (%)IOBs: " + percent_IOB + ", Used IPADs: " + used_ipads + ", (%)IPADs: " + percent_ipads + ", Used OPADs: " + used_opads + ", (%) OPADs: " + percent_opads + ", GTs: " + gts)
                    if(g == 2):
                        line = fp.readline()
                    line = fp.readline()
            line = fp.readline()
```
